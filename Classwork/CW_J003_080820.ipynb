```
def factorial_recusrsion(n):
  factor = 1
  if(n==0):
      return factor
  else:
      return (n*factorial_recursion(n-1))


def factorial_simple(n):
  factor = 1
  for i in range(1,n+1):
    factor = factor*i
  return factor
```

```
import time
start_time = time.time()
print(factorial_simple(5))
print(time.time() - start_time)
%timeit (factorial_simple(5))
```
